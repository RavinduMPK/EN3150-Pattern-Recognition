```python
import numpy as np
import matplotlib . pyplot as plt
from sklearn . datasets import make_circles
# Generate data with make_circles
np. random . seed (5)
X, y = make_circles ( n_samples =500 , factor =0.3 , noise =0.1)
plt.scatter (X [:, 0], X [:, 1], c=y, cmap=plt.cm.Paired)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def mapping(x1,x2):

    # Apply the mapping  to project the data points to the three-dimensional space
    x1_proj = x1
    x2_proj = x2
    x3_proj = x1**2 + x2**2

    return x1_proj, x2_proj, x3_proj

x1_proj, x2_proj, x3_proj = mapping(X[:,0], X[:,1])

# Create a 3D plot to visualize the projected set
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x1_proj, x2_proj, x3_proj, c=y, cmap=plt.cm.Paired)

# Set labels for the axes
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('x3')

# Show the plot
plt.show()
```

```python
# Use another mapping function to project the data points to the three-dimensional space
import numpy as np
import matplotlib.pyplot as plt

def mapping_2(x1,x2):

    # Apply the mapping  to project the data points to the three-dimensional space
    x1_proj = x1**2
    x2_proj = x2**2
    x3_proj = x1*x2

    return x1_proj, x2_proj, x3_proj

x1_proj_1, x2_proj_1, x3_proj_1 = mapping_2(X[:,0], X[:,1])

# Create a 3D plot to visualize the projected set
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x1_proj_1, x2_proj_1, x3_proj_1, c=y, cmap=plt.cm.Paired)

# Set labels for the axes
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('x3')

# Show the plot
plt.show()
```

```python
# Data without mapping

from sklearn.model_selection import train_test_split
import sklearn.svm as svm
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)

# Train a linear SVM classifier
clf = svm.SVC(kernel='linear')
clf.fit(X_train, y_train)

# Predict the labels of the test set
y_pred = clf.predict(X_test)

# Compute the evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Print the evaluation metrics
print('Accuracy: %.2f' % accuracy)
print('Precision: %.2f' % precision)
print('Recall: %.2f' % recall)
print('F1: %.2f' % f1)
```

```python
# Data with mapping (x1_proj, x2_proj, x3_proj)

from sklearn.model_selection import train_test_split
import sklearn.svm as svm
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(np.c_[x1_proj, x2_proj, x3_proj], y, test_size=0.3, random_state=5)

# Train a linear SVM classifier
clf = svm.SVC(kernel='linear')
clf.fit(X_train, y_train)

# Predict the labels of the test set
y_pred = clf.predict(X_test)

# Compute the evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Print the evaluation metrics
print('Accuracy: %.2f' % accuracy)
print('Precision: %.2f' % precision)
print('Recall: %.2f' % recall)
print('F1: %.2f' % f1)
```
