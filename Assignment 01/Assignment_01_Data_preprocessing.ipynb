```python
import numpy as np
import matplotlib . pyplot as plt
import pandas as pd

def generate_signal ( signal_length , num_nonzero ):
    signal = np. zeros ( signal_length )
    nonzero_indices = np. random . choice ( signal_length ,num_nonzero , replace = False )
    nonzero_values = 50* np. random . randn ( num_nonzero )
    signal [ nonzero_indices ] = nonzero_values
    return signal

# Data generation
signal_length = 100 # Total length of the signal
num_nonzero = 10
your_index_no =200488 # Enter your index number ( without English letter and without leading zeros )
signal = generate_signal ( signal_length , num_nonzero )
signal [10] = ( your_index_no % 10) *10 + 10
if your_index_no % 10 == 0:
    signal [10] = np. random . randn (1) + 30

signal = signal . reshape ( signal_length ,1)

# plotting
plt . figure ( figsize =(15 ,5) )
plt . subplot (1, 1, 1)
plt . title (" Data ")
plt . stem ( signal )
```

```python
print(pd.DataFrame(signal, columns=['signal']).describe())
```

```python
import sklearn.preprocessing as pre

# Apply MaxAbsScaler to signal
max_abs_scaler = pre. MaxAbsScaler ()
signal_max_abs = max_abs_scaler . fit_transform ( signal )

# Visualize the data before and after normalization (Stem plots)
plt . figure ( figsize =(15 ,5) )
plt . subplot (1, 2, 1)
plt . title (" Data before normalization ")
plt . stem ( signal )
plt . subplot (1, 2, 2)
plt . title (" Data after MaxAbs normalization ")
plt . stem ( signal_max_abs )

print(pd.DataFrame(signal_max_abs, columns=['signal']).describe())
```

```python
# Apply MinMaxScaler to signal

def min_max_scaler ( signal ):
    min_val = np. min ( signal )
    max_val = np. max ( signal )
    signal_min_max = ( signal - min_val ) / ( max_val - min_val )
    return signal_min_max

signal_min_max = min_max_scaler ( signal )

# Visualize the data before and after normalization (Stem plots)
plt . figure ( figsize =(15 ,5) )
plt . subplot (1, 2, 1)
plt . title (" Data before normalization ")
plt . stem ( signal )
plt . subplot (1, 2, 2)
plt . title (" Data after MinMax normalization ")
plt . stem ( signal_min_max )

print(pd.DataFrame(signal_min_max, columns=['signal']).describe())
```

```python
# Apply Standard Normalization to signal
def standard_normalization ( signal ):
    mean = np. mean ( signal )
    std = np. std ( signal )
    signal_standard = ( signal - mean ) / std
    return signal_standard

signal_standard = standard_normalization ( signal )

# Visualize the data before and after normalization (Stem plots)
plt . figure ( figsize =(15 ,5) )
plt . subplot (1, 2, 1)
plt . title (" Data before normalization ")
plt . stem ( signal )
plt . subplot (1, 2, 2)
plt . title (" Data after Standard normalization ")
plt . stem ( signal_standard )

print(pd.DataFrame(signal_standard, columns=['signal']).describe())
```

```python
# None-zero elements of signal before and after normalization

# MaxAbsScaler
nonzero_signal = signal [ signal != 0]
nonzero_signal_max_abs = signal_max_abs [ signal_max_abs != 0]
plt . figure ( figsize =(15 ,5) )
plt . subplot (1, 2, 1)
plt . title (" None - zero elements of signal before normalization ")
plt . stem ( nonzero_signal )
plt . subplot (1, 2, 2)
plt . title (" None - zero elements of signal after MaxAbs normalization ")
plt . stem ( nonzero_signal_max_abs )

# Print the count of non-zero elements in signal before and after normalization
print ( " Number of non - zero elements in signal before normalization : " , len ( nonzero_signal ))
print ( " Number of non - zero elements in signal after MaxAbs normalization : " , len ( nonzero_signal_max_abs ))


# MinMaxScaler
nonzero_signal = signal [ signal != 0]
nonzero_signal_min_max = signal_min_max [ signal_min_max != 0]
plt . figure ( figsize =(15 ,5) )
plt . subplot (1, 2, 1)
plt . title (" None - zero elements of signal before normalization ")
plt . stem ( nonzero_signal )
plt . subplot (1, 2, 2)
plt . title (" None - zero elements of signal after MinMax normalization ")
plt . stem ( nonzero_signal_min_max )

# Print the count of non-zero elements in signal before and after normalization
print ( " Number of non - zero elements in signal before normalization : " , len ( nonzero_signal ))
print ( " Number of non - zero elements in signal after MinMax normalization : " , len ( nonzero_signal_min_max ))


# Standard Normalization
nonzero_signal = signal [ signal != 0]
nonzero_signal_standard = signal_standard [ signal_standard != 0]
plt . figure ( figsize =(15 ,5) )
plt . subplot (1, 2, 1)
plt . title (" None - zero elements of signal before normalization ")
plt . stem ( nonzero_signal )
plt . subplot (1, 2, 2)
plt . title (" None - zero elements of signal after Standard normalization ")
plt . stem ( nonzero_signal_standard )

# Print the count of non-zero elements in signal before and after normalization
print ( " Number of non - zero elements in signal before normalization : " , len ( nonzero_signal ))
print ( " Number of non - zero elements in signal after Standard normalization : " , len ( nonzero_signal_standard ))


```
